```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s_initial = 100000
r_initial = 0;
i_initial = 10;

beta = 0.0000001
gamma = 0.0001

steps = 10000
```

```python
def sir_next(s, i, r, beta, gamma):
    s_diff = i * beta * s
    r_diff = i * gamma 
    r_new = r + r_diff
    s_new = np.max([0, s - s_diff])
    i_new = np.max([0, i + s_diff - r_diff])
    return s_new, i_new, r_new

    
```

```python
s = s_initial;
r = r_initial;
i = i_initial;
data = []
for t in range(steps):
    s_new, i_new, r_new = sir_next(s,i,r, beta, gamma)
    data.append( (s_new, i_new, r_new))
    s,i,r = s_new, i_new, r_new
    
```

```python
plt.plot(
    range(len(data)),
    [item[0] for item in data], 'r',
    [item[1] for item in data], 'b',
    [item[2] for item in data], 'g'
)
plt.show()
```
