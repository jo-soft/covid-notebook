```python
from dataSrc import CovidDataSrc
cds = CovidDataSrc()
```

```python
cds.get_covid_data_for_lk('Münster')
```

```python
ds_ms = cds.get_covid_data_for_lk_with_neighbours('Münster')
```

```python
ds_ms.div(ds_ms.max()).plot(figsize=(18,9))
```

```python
ds_ms_rolling = ds_ms.div(ds_ms.max()).rolling(window=7, closed='both', win_type='gaussian')
```

```python
ds_ms_rolling.mean(std=3).plot(figsize=(18,8))
```

```python
ds_ms.div(ds_ms.max()).rolling(14).mean(std=7).plot(figsize=(18,8))
```

```python
ds_ms.div(ds_ms.max()).corr()
```
